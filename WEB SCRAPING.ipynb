Grabbing the title of a page

```python
import requests
```

```python
result= requests.get("http://www.example.com")
```

```python
type(result)
```

```python
result.text
```

```python
#in order to parse through the sgtring above we will have to use BeautifulSoup
```

```python
import bs4
```

```python
soup= bs4.BeautifulSoup(result.text,"lxml")
```

```python
soup
```

```python
#now to grap the title 'Example domain'(<title>Example Domain</title>)...
```

```python
soup.select('title') #to grab the title tag...
```

```python
soup.select('title')[0].getText() #to grab the text in between the tags...
```

```python
soup.select('p')
```

```python
soup.select('p')[0].getText()
```
